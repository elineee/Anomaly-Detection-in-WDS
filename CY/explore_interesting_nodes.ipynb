Explore dependencies dict to check what are the interesting nodes

```python
import pickle


with open('.//dict_dependencies_significant_nodes.pkl', 'rb') as f:
    loaded_dict = pickle.load(f)
print(loaded_dict)

# get nodes affected by a contamination at dist423
for key, value in loaded_dict.items():
    if key == "dist423":
        print(value)

# Find the keys that have "dist1903" in their values
keys = []
for key, value in loaded_dict.items():
    for element in value: 
        if element == "dist1903":
            keys.append(key)
print(keys)
```
